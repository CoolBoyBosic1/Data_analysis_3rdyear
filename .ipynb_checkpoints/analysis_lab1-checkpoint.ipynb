### 2) Характеристики положення центра
 Середнє арифм. (mean)
 
 Середнє геометричне (geometric mean) — визначене для додатних значень
 
 Середнє гармонійне (harmonic mean) — визначене для додатних значень
 
 Мода (mode) 
 
 Медіана (median)

```python
def safe_geo_mean(x):
    x = np.asarray(x, float)
    x = x[x > 0]
    return np.exp(np.log(x).mean()) if len(x) else np.nan

def safe_harm_mean(x):
    x = np.asarray(x, float)
    x = x[x > 0]
    return stats.hmean(x) if len(x) else np.nan

def hist_mode(x, bins="auto"):
    # модальна точка як центр біна з найбільшою щільністю
    counts, edges = np.histogram(x[~np.isnan(x)], bins=bins)
    if counts.size == 0:
        return np.nan
    i = counts.argmax()
    return (edges[i] + edges[i+1]) / 2

center_table = pd.DataFrame({
    "mean": bf.mean(),
    "median": bf.median(),
    "geometric_mean": bf.apply(safe_geo_mean, axis=0),
    "harmonic_mean": bf.apply(safe_harm_mean, axis=0),
    "mode": bf.mode().iloc[0]
})
center_table
```

### 6) Перевірка стохастичності
 
 Критерій серій відносно **медіани** (Вальд–Вольфовіц)

 Критерій **зростаючих/спадаючих** серій
 
 **Критерій Аббе** (квадрати послідовних різниць)

```python
# %% 3.1 runs by median / runs up-down / Abbe — stable numpy versions
def runs_median_test(x):
    x = pd.Series(x).to_numpy()
    if x.size < 10: return np.nan, np.nan
    med = np.median(x)
    s = (x > med).astype(int)
    s = s[x != med]  # прибрати рівні медіані
    if s.size < 10 or s.sum()==0 or s.sum()==s.size:
        return np.nan, np.nan
    runs = 1 + np.sum(np.diff(s) != 0)
    n1 = s.sum(); n2 = s.size - n1
    mu = 2*n1*n2/(n1+n2) + 1
    var = (2*n1*n2*(2*n1*n2 - n1 - n2))/(((n1+n2)**2)*(n1+n2-1))
    z = (runs - mu)/np.sqrt(var)
    p = 2*(1 - stats.norm.cdf(abs(z)))
    return z, p

def runs_up_down_test(x):
    x = pd.Series(x).to_numpy()
    if x.size < 10: return np.nan, np.nan
    d = np.sign(np.diff(x))
    d = d[d!=0]
    if d.size < 8: return np.nan, np.nan
    runs = 1 + np.sum(np.diff(d) != 0)
    n1 = (d>0).sum(); n2 = (d<0).sum()
    if n1==0 or n2==0: return np.nan, np.nan
    mu = 2*n1*n2/(n1+n2) + 1
    var = (2*n1*n2*(2*n1*n2 - n1 - n2))/(((n1+n2)**2)*(n1+n2-1))
    z = (runs - mu)/np.sqrt(var)
    p = 2*(1 - stats.norm.cdf(abs(z)))
    return z, p

def abbe_test(x):
    x = pd.Series(x).to_numpy()
    n = x.size
    if n < 10: return np.nan, np.nan, np.nan
    num = np.sum(np.diff(x)**2)/(n-1)
    den = np.var(x, ddof=1)
    if den == 0: return np.nan, np.nan, np.nan
    gamma = num/(2*den)
    z = (gamma - 1)/np.sqrt(0.5/n)       
    p = 2*(1 - stats.norm.cdf(abs(z)))
    return gamma, z, p

tests6 = []
for c in num_cols:
    z1,p1 = runs_median_test(bf[c])
    z2,p2 = runs_up_down_test(bf[c])
    g,z3,p3 = abbe_test(bf[c])
    tests6.append([c,z1,p1,z2,p2,g,z3,p3])

tests6 = pd.DataFrame(tests6, columns=[
    "var","z_runs_median","p_runs_median","z_runs_updown","p_runs_updown",
    "gamma_abbe","z_abbe","p_abbe"
]).set_index("var")

display(tests6.round(4).style.set_caption("Стохастичність: серії (медіана), серії, Аббе"))
```

### 7) Рангові критерії однорідності
 Якщо є явні групи (наприклад, стовпець **group** у bf), тестуємо однорідність:

 Mann–Whitney (2 групи)

 Kruskal–Wallis (>=3 груп)

 Якщо груп немає — робимо "split-half" перевірку (рандомно ділимо на 2 підвибірки однакового розміру).


```python
# %%
def rank_homogeneity(df, group_col=None):
    out = {}
    if group_col and group_col in df.columns:
        g = df[group_col].dropna().astype("category")
        levels = g.cat.categories.tolist()
        idxs = [df.index[g==lv] for lv in levels]
        if len(levels)==2:
            res={}
            for c in df.select_dtypes(include=[np.number]).columns:
                x1 = df.loc[idxs[0],c].dropna(); x2 = df.loc[idxs[1],c].dropna()
                if len(x1) and len(x2):
                    U,p = stats.mannwhitneyu(x1,x2,alternative="two-sided")
                    res[c]={"U":U,"p":p}
            out["mannwhitney"] = pd.DataFrame(res).T.sort_values("p")
        elif len(levels)>=3:
            res={}
            for c in df.select_dtypes(include=[np.number]).columns:
                arrays=[df.loc[i,c].dropna() for i in idxs]
                if all(len(a) for a in arrays):
                    H,p = stats.kruskal(*arrays)
                    res[c]={"H":H,"p":p,"k":len(levels)}
            out["kruskal"] = pd.DataFrame(res).T.sort_values("p")
    else:
        res={}
        for c in df.select_dtypes(include=[np.number]).columns:
            x = df[c].dropna().values
            if len(x)>=8:
                idx = rng.permutation(len(x))
                half=len(x)//2
                U,p = stats.mannwhitneyu(x[idx[:half]], x[idx[half:]], alternative="two-sided")
                res[c]={"U_split":U,"p_split":p,"n1":half,"n2":len(x)-half}
        out["split_half"] = pd.DataFrame(res).T.sort_values("p_split")
    return out

rank7 = rank_homogeneity(bf)   
for name,tab in rank7.items():
    display(tab.round(4).style.set_caption(f"Рангові тести однорідності: {name}"))
```

### 8) Перевірка симетрії розподілу ранговими критеріями

 Wilcoxon signed-rank на відхилення від медіани (H0: симетрія)

```python
def symmetry_wilcoxon(x):
    x = pd.Series(x).dropna().values
    med = np.median(x)
    d = x - med
    d = d[d!=0]
    if len(d)<10: return np.nan, np.nan
    stat,p = stats.wilcoxon(d, alternative="two-sided", zero_method="wilcox")
    return stat,p

sym8 = bf.apply(lambda s: pd.Series(symmetry_wilcoxon(s), index=["W","p_symmetry"]))
display(sym8.round(4).style.set_caption("Симетрія розподілу (рангово, Wilcoxon)"))
```

### 9) Виявлення аномалій та чистка (Grubbs + узагальнений ESD)

 Знаходимо викиди по кожній змінній: одноточкові - Grubbs, множинні - узагальнений ESD (Tietjen–Moore) з розумним максимумом 𝑘(беремо 10 ≈ 4% від n=252). Об’єднуємо індекси -> формуємо CLEAN, переобчислюємо ключові таблиці та heatmap.

```python
# %% 9) Outliers → CLEAN → recalc (мінімально)
num_cols = bf.select_dtypes(include=[np.number]).columns.tolist()
ALPHA = 0.05
MAX_ESD = 10  # ~ 4% від n=252

union_idx = set()

# 9.1 Grubbs (по одному викиду)
for c in num_cols:
    s = bf[c].dropna()
    if s.size >= 3 and s.std(ddof=1) > 0:
        z = (s.sub(s.mean()).abs() / s.std(ddof=1))
        i = z.idxmax(); G = z.loc[i]
        n = s.size
        t = stats.t.ppf(1 - ALPHA/(2*n), n-2)
        Gcrit = (n-1)/np.sqrt(n) * np.sqrt(t**2/(n-2+t**2))
        if np.isfinite(G) and G > Gcrit:
            union_idx.add(i)

# 9.3 Узагальнений ESD (Tietjen–Moore), k <= MAX_ESD
for c in num_cols:
    s = bf[c].dropna().copy()
    removed = 0
    while s.size >= 5 and removed < MAX_ESD:
        if s.std(ddof=1) == 0:
            break
        z = (s.sub(s.mean()).abs() / s.std(ddof=1))
        j = z.idxmax(); R = z.loc[j]
        n_curr = s.size
        p = 1 - ALPHA/(2*n_curr)
        t = stats.t.ppf(p, n_curr-2)
        lam = (n_curr-1)*t / np.sqrt((n_curr-2+t**2)*(n_curr+1))
        if np.isfinite(R) and np.isfinite(lam) and R > lam:
            union_idx.add(j)
            s = s.drop(j)
            removed += 1
        else:
            break

print(f"Знайдено підозрілих вимірів (об'єднано): {len(union_idx)}")

bf_clean = bf.drop(index=list(union_idx)).reset_index(drop=True)

# повторний центр/розсіювання (CLEAN)
center_table_clean = pd.DataFrame({
    "mean": bf_clean[num_cols].mean(),
    "median": bf_clean[num_cols].median(),
    "geometric_mean": bf_clean[num_cols].apply(lambda x: np.exp(np.log(x[x>0]).mean()) if (x>0).any() else np.nan),
    "harmonic_mean": bf_clean[num_cols].apply(lambda x: stats.hmean(x[x>0]) if (x>0).any() else np.nan),
    "mode_real": bf_clean[num_cols].mode().iloc[0]
})

spread_table_clean = pd.DataFrame({
    "variance": bf_clean[num_cols].var(ddof=1),
    "std": bf_clean[num_cols].std(ddof=1),
})
spread_table_clean["conc_low"]  = center_table_clean["mean"] - 3*spread_table_clean["std"]
spread_table_clean["conc_high"] = center_table_clean["mean"] + 3*spread_table_clean["std"]

display(center_table_clean.round(4).style.set_caption("Центральні характеристики (CLEAN)"))
display(spread_table_clean.round(4).style.set_caption("Розсіювання + [x̄±3s] (CLEAN)"))

# heatmap (CLEAN, загальний випадок)
C_clean = bf_clean[num_cols].corr()
plt.figure(figsize=(8.2, 7))
im = plt.imshow(C_clean, vmin=-1, vmax=1)
plt.colorbar(im, fraction=0.046, pad=0.04)
plt.xticks(range(len(num_cols)), num_cols, rotation=90, fontsize=8)
plt.yticks(range(len(num_cols)), num_cols, fontsize=8)
for i in range(C_clean.shape[0]):
    for j in range(C_clean.shape[1]):
        plt.text(j, i, f"{C_clean.iat[i,j]:.2f}", ha="center", va="center", fontsize=7, color="black")
plt.title("Correlation Heatmap (CLEAN)")
plt.tight_layout(); plt.show()
```

### 10) Визначення характеристики статистичного зв'язку

 Через велику кількість повторів (ties) у більшості змінних - основна парна міра для наступного етапу Kendall. Для прозорості показуємо частку ties.

```python
num_cols = bf_clean.select_dtypes(include=[np.number]).columns.tolist()

ties_rows = []
C2 = lambda k: k*(k-1)/2
for c in num_cols:
    x = bf_clean[c].dropna().to_numpy(); n = x.size
    if n == 0:
        ties_rows.append({"var": c, "n": 0, "n_unique": 0, "ties_obs": np.nan, "ties_pairs": np.nan, "unique_ratio": np.nan})
    else:
        vals, cnts = np.unique(x, return_counts=True)
        ties_rows.append({
            "var": c, "n": n, "n_unique": len(vals),
            "ties_obs": cnts[cnts>1].sum()/n,
            "ties_pairs": (np.sum(C2(cnts))/C2(n)) if n>=2 else 0.0,
            "unique_ratio": len(vals)/n
        })

ties_table = pd.DataFrame(ties_rows).sort_values("ties_pairs", ascending=False).reset_index(drop=True)
display(ties_table.round(4).style.set_caption("Частка зв'язаних значень (CLEAN): obs vs pairs + %unique"))
```

```python
#тест на нормальність розподілу Колмогорова-Смірнова

alpha = 0.05

for col in num_cols:
    data = (bf_clean[col] - bf_clean[col].mean()) / bf_clean[col].std()
    _, p_value = stats.kstest(data, 'norm')
    if p_value > alpha:
        display(f"{col}: Нормальний розподіл(p={p_value:.4f}) -> Використовую Пірсона")
    else:
        display(f"{col}: НЕ нормальний розподіл(p={p_value:.4f}) -> Використовую Спірмена/Кендалла")
```

У більшості змінних висока зв’язаність (наприклад, ties_obs 0.24–0.97; ties_pairs до 0.034) -> дані дискретизовані/округлені й не гарантують нормальності. Тому для парного аналізу беремо Kendall: він стійкий до викидів, інваріантний до масштабування та коректно враховує ties (на відміну від pearson та spearman). Spearman/Pearson залишаємо довідково для інтерпретації (монотонність/лінійність) + (ties_obs для останніх 5 змінних доволі низький) 

При цьому тест Колмогорова-Смірнова показав що всі наші змінні нормально розподілені, тому можна буде використати в третьому етапі коефіцієнт детермінації та використати кореляцію пірсона та кендала для аналізу істотності парних статистичних зв'язків

# Аналіз істотності парних статистичних зв'язків

### Обчислення всіх пар (Kendall — основна) + p, FDR

```python
df_pairs = bf_clean
num_cols = df_pairs.select_dtypes(include=[np.number]).columns.tolist()

rows = []
for a, b in combinations(num_cols, 2):
    x = df_pairs[a].dropna(); y = df_pairs[b].dropna()
    idx = x.index.intersection(y.index)
    x = x.loc[idx].to_numpy(); y = y.loc[idx].to_numpy()
    if x.size >= 3:
        tau, p_tau = stats.kendalltau(x, y, method='auto')
        rho, p_rho = stats.spearmanr(x, y)
        try: r, p_r = stats.pearsonr(x, y)
        except Exception: r, p_r = np.nan, np.nan
        rows.append({"var1":a,"var2":b,"n_pair":x.size,
                     "kendall_tau":tau,"p_tau":p_tau,     
                     "spearman_rho":rho,"p_rho":p_rho,
                     "pearson_r":r,"p_r":p_r})

pairs = pd.DataFrame(rows)

# Benjamini–Hochberg (правильна реалізація з cummin)
p = pairs["p_tau"].to_numpy()
m = (~np.isnan(p)).sum()
order = np.argsort(p, kind="mergesort")
p_sorted = p[order]
adj = p_sorted * m / (np.arange(1, m+1))
adj = np.minimum.accumulate(adj[::-1])[::-1]
p_adj = np.full_like(p, np.nan, dtype=float)
p_adj[order] = adj
pairs["p_tau_fdr"] = np.clip(p_adj, 0, 1)

# Сортування за істотністю (менший p_tau → істотніше)
pairs_sorted = pairs.sort_values(["p_tau","p_tau_fdr","kendall_tau"], ascending=[True,True,False]).reset_index(drop=True)

cols = ["var1","var2","n_pair","kendall_tau","p_tau","p_tau_fdr","spearman_rho","p_rho","pearson_r","p_r"]
display(pairs_sorted[cols].round(4).style.set_caption(
    "Парні зв’язки: Kendall τ (значення + p + FDR). Spearman/Pearson — довідково."
))
```

### Впорядкований список пар (спадання істотності)

```python
display(
    pairs_sorted.loc[:,["var1","var2","kendall_tau","p_tau","p_tau_fdr","n_pair"]]
    .round(4)
    .style.set_caption("Усі пари за істотністю (Kendall tau)")
)
```

### Висновок

```python
sig = pairs_sorted[pairs_sorted["p_tau_fdr"] <= 0.05].copy()
print(f"Значущих пар за Kendall τ (FDR≤0.05): {len(sig)}")

if not sig.empty:
    pos = sig.sort_values("kendall_tau", ascending=False).head(5)
    neg = sig.sort_values("kendall_tau", ascending=True).head(5)
    print("\nТоп + зв’язки (tau):")
    for _,r in pos.iterrows():
        print(f"  {r.var1} — {r.var2}: tau ={r.kendall_tau:.3f}, p={r.p_tau:.2e}, q(FDR)={r.p_tau_fdr:.2e}")
    print("\nТоп − зв’язки (tau):")
    for _,r in neg.iterrows():
        print(f"  {r.var1} — {r.var2}: tau ={r.kendall_tau:.3f}, p={r.p_tau:.2e}, q(FDR)={r.p_tau_fdr:.2e}")
```

Загалом картина узгоджена: Kendall  дає той самий топ, що й spearman, pearson, і саме його,було правильно взяти через велику зв’язаність даних. Для майже лінійних(в лапках)) пар з відносно меншими ties (на кшталт Weight–Hip, Chest–Abdomen, BodyFat–Abdomen, Weight–Chest) Pearson ~ Spearman ~ Kendall (|r|,|p| >= ~0.80), тож pearson тут придатний для лінійної інтерпретації (силу/нахил зв’язку). У змінних з більшою дискретністю (Age, Wrist, Height, Ankle тощо) рангові міри (Kendall/ Spearman) надійніші; Pearson легко «просідає» через ties/викиди (пари з Age мають малі |r|, вищі p).
Слабкі зв’язки (низ таблиці, зокрема з Age або Height–Abdomen/Chest) залишаються неістотними навіть без FDR або мають дуже малий ефект - їх інтерпретуємо з обережністю

Жир vs щільність тіла: BodyFat <-> Density -- майже дзеркало (дуже сильна обернена залежність): більше жиру => менша щільність.

Вага та обхвати: Weight тісно пов’язана з Hip, Chest, Abdomen, Thigh, Knee, Forearm, Biceps ⇒ збільшення маси зазвичай іде разом із більшими обхватами.

Жир і «центральні» обхвати: BodyFat найсильніше йде з Abdomen (далі Chest, Hip) => абдомінальний обхват - найкращий простий проксі відсотка жиру.

Кластери тіла:
   Верхній торс: Chest–Abdomen–Hip сильно пов’язані між собою та з Weight/BodyFat.
   Нижня частина: Hip–Thigh–Knee–Ankle утворюють окремий «розмірний» кластер.
   Руки: Biceps–Forearm–Wrist координуються; Forearm краще відбиває м’язову/масову складову, Wrist - радше «розмір кістки/фрейм».

Шия як маркер ожиріння: Neck корелює з Chest/Abdomen/Hip/Weight -> може слугувати швидким сурогатом загальної масивності/ожиріння.

Зріст (Height): слабко пов’язаний із BodyFat та більшістю обхватів -> вищі люди не обов’язково мають більший % жиру; зріст радше задає масштаб скелета.

Вік (Age): здебільшого слабкі або неістотні парні зв’язки -> в цій вибірці вік сам по собі не пояснює різницю в жирових/розмірних показниках.

Висновок: якщо треба швидко оцінити жир, найінформативніші - Abdomen (перший), далі Chest/Hip; для загальної масивності - Weight + Hip/Chest; для м’язової компоненти верхніх кінцівок — Forearm/Biceps.

# Аналіз істотності множинних статистичних зв'язків

```python
df = bf_clean.select_dtypes(include=[np.number]).dropna()
cols = df.columns.tolist(); n, p = df.shape

# кореляційна матриця (Spearman)
R = df.corr(method="spearman").to_numpy()

# множинні метрики для кожної змінної як Y
rows = []
for j, y in enumerate(cols):
    idx = [i for i in range(p) if i != j]
    Rxx = R[np.ix_(idx, idx)]
    r_yx = R[j, idx]
    R2 = float(r_yx @ np.linalg.pinv(Rxx) @ r_yx.T)
    k = p - 1
    F = (R2 / k) / ((1 - R2) / (n - k - 1))
    p_overall = 1 - stats.f.cdf(F, k, n - k - 1)
    rows.append({"var": y, "n": n, "k": k, "R2": R2, "R2_adj": 1 - (1 - R2) * (n - 1) / (n - k - 1), "p_overall": p_overall})

multi = pd.DataFrame(rows)

# BH-FDR для p_overall + впорядкування
pvals = multi["p_overall"].to_numpy(); m = np.isfinite(pvals).sum()
order = np.argsort(pvals, kind="mergesort"); ps = pvals[order]
q = ps * m / (np.arange(1, m + 1)); q = np.minimum.accumulate(q[::-1])[::-1]
qq = np.full_like(pvals, np.nan, dtype=float); qq[order] = np.clip(q, 0, 1)
multi["q_overall"] = qq
multi_sorted = multi.sort_values(["q_overall", "p_overall", "R2_adj", "R2"], ascending=[True, True, False, False]).reset_index(drop=True)
```

```python
# вивід таблиць 
display(
    multi_sorted.loc[:, ["var","R2","R2_adj","p_overall","q_overall","n","k"]]
    .round(4)
    .style.set_caption("Множинні зв’язки (Spearman): R2 / R2_adj + p(F) + q(FDR)")
)

display(
    multi_sorted.loc[:, ["var","q_overall"]]
    .round(4)
    .style.set_caption("Послідовність змінних за спаданням істотності (q_overall)")
)
```

```python
# короткі висновки 
alpha = 0.05
sig = multi_sorted[multi_sorted["q_overall"] <= alpha]
print(f"Істотних при FDR≤{alpha}: {sig.shape[0]}/{multi_sorted.shape[0]}")
print("Топ-5 за R2_adj:")
print(sig.sort_values("R2_adj", ascending=False).loc[:, ["var","R2_adj"]].head(5).round(4).to_string(index=False))
```
